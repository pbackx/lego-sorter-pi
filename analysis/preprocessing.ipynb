```python
import cv2 as cv
from os import listdir
from matplotlib import pyplot as plt
import random
```

```python
images = [f'..\\data\\{file}' for file in listdir('..\\data') if file.endswith('.jpg') and not file.startswith('reference')]
```

```python
image = cv.imread(images[0])
```

```python
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
```

```python
image.shape
```

```python
margin = int((640-480)/2)
cropped = image[:, margin:-margin]
plt.imshow(cv.cvtColor(cropped, cv.COLOR_BGR2RGB))
```

```python
resized = cv.resize(cropped, (256,256), interpolation=cv.INTER_AREA)
plt.imshow(cv.cvtColor(resized, cv.COLOR_BGR2RGB))
```
